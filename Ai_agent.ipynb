```python
import subprocess
import os
import time
import requests

# Set environment variables for Ollama
os.environ['OLLAMA_HOST'] = '0.0.0.0'
os.environ['OLLAMA_PORTS'] = '11434' # Explicitly set port

ollama_url = 'http://0.0.0.0:11434'

# Check if Ollama is already running. If it is, do not attempt to start it again.
try:
    response = requests.get(ollama_url, timeout=1)
    if response.status_code == 200:
        print("Ollama server is already running.")
    else:
        raise requests.exceptions.ConnectionError
except requests.exceptions.ConnectionError:
    print("Ollama server is not running, starting it now...")
    # Use nohup to ensure the process continues after the Colab cell finishes
    # Redirect stdout and stderr to a log file
    command = "nohup /usr/local/bin/ollama serve > ollama.log 2>&1 &"
    subprocess.run(command, shell=True, check=True)

    # Wait for the server to become available
    print("Waiting for Ollama server to start...")
    max_retries = 30 # Wait up to 30 seconds
    for i in range(max_retries):
        try:
            response = requests.get(ollama_url, timeout=1)
            if response.status_code == 200:
                print(f"Ollama server started successfully after {i+1} seconds.")
                break
        except requests.exceptions.ConnectionError:
            pass
        time.sleep(1)
    else:
        print("Ollama server failed to start within the expected time.")
        print("Check 'ollama.log' for more details.")
```

```python
from langchain_ollama import OllamaLLM

llm = OllamaLLM(model="llama3")

def run_llm(prompt):
    response = llm.invoke(prompt)
    return response
```

```python
def planner_agent(app_idea):
    prompt = f"""
    You are a senior software architect.

    Break down the following app idea into clear development steps:
    {app_idea}

    Include:
    - Features
    - Tech stack
    - Step-by-step implementation plan
    """
    return run_llm(prompt)
```

```python
def developer_agent(plan):
    prompt = f"""
    You are a Python developer.

    Based on this plan:
    {plan}

    Write clean, modular, production-ready Python code.
    """
    return run_llm(prompt)
```

```python
def tester_agent(code):
    prompt = f"""
    You are a software tester.

    Analyze this code:
    {code}

    Identify:
    - Bugs
    - Edge cases
    - Improvements
    """
    return run_llm(prompt)
```

```python
def reviewer_agent(code):
    prompt = f"""
    You are a senior code reviewer.

    Improve this code:
    {code}

    Focus on:
    - Readability
    - Performance
    - Best practices
    """
    return run_llm(prompt)
```

```python
from rich import print

def build_app(app_idea):
    print("[bold cyan]Step 1: Planning...[/bold cyan]")
    plan = planner_agent(app_idea)
    print(plan)

    print("\n[bold green]Step 2: Developing...[/bold green]")
    code = developer_agent(plan)
    print(code)

    print("\n[bold yellow]Step 3: Testing...[/bold yellow]")
    test_report = tester_agent(code)
    print(test_report)

    print("\n[bold magenta]Step 4: Reviewing...[/bold magenta]")
    final_code = reviewer_agent(code)
    print(final_code)

    return final_code

if __name__ == "__main__":
    idea = "Build a simple REST API for a todo app using FastAPI"
    final_output = build_app(idea)
```
